```python
#FIXME
#Parte 2
#Punto 4
df_new = df.loc[(df['type']=='Movie') & (df['year_added']>2018)]
df_new.head()
```

```python
#Punto 5
df_new = df.loc[df['title'].str.lower().str.contains('love')]
df_new.head()
```

```python
#Punto 5
movies_df = df.loc[df['type'] == 'Movie'].copy()
movies_df['duration_minutes'] = movies_df['duration'].str.extract('(\d+) min').astype(float)
display(movies_df.head())
```

```python
#Punto 6
genero_df = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')
genero_df.head()
```

```python
#Punto 7
genero_df['listed_in'].value_counts().head(10)
```

```python
#Punto 8
movies_df['long_content'] = movies_df['duration_minutes'].where(movies_df['duration_minutes'] > 120, 'short')
movies_df['long_content'] = movies_df['long_content'].mask(movies_df['duration_minutes'] > 120, 'long')
movies_df.head()
```

```python
#Punto 9
#Primero agrego la columna duration_minutes al dataframe
df['duration_minutes'] = df['duration'].where(df['type'] == 'Movie').str.extract(r'(\d+)').astype(float)
#Luego saco la informacion
df_minute= df.loc[(df['type']=='Movie') & (df['duration_minutes']>100) & (df['rating']=='R') & (df['country']=='United States')]
df_minute.head()
```

```python
#Punto 10 Utilizar
top10_df=movies_df.sort_values('duration_minutes',ascending=False).head(10)
top10_df.style.bar(subset=['duration_minutes'], color='yellow', align='mid')
```

```python
#Punto 11
df.dropna(subset=['listed_in','rating'], inplace=True)
df['genre_list'] = df['listed_in'].str.split(',')
# Aplico explode para que cada genero tenga su propia fila
df_exploded=df.explode('genre_list')
df_exploded.rename(columns={'genre_list': 'genre'}, inplace=True)
top_combinacion=df_exploded.value_counts(subset=['genre', 'rating'])
top_combinacion.head()
```

```python
#Punto 12 Bonus
tduplicado_df=df[df.duplicated(subset=['title'], keep=False)]
tduplicado_df.head()
```

Se puede ver que no genera ninguna pelicula, por tanto no existen peliculas con el mismo nombre pero con distinto año de lanzamiento.
